```python
import numpy as np
import os.path
```

```python
train_path = 'train.txt'
f = open(train_path,'r')
o = open('train2.txt','a')
lines = f.readlines()
f.close()
count = 0
for line in lines:
    tmp = line.split(' ')
    if os.path.isfile(tmp[0]):
        print(tmp[0] + " File exist")
        o.write(line)
    else:
        count += 1
        print (tmp[0] + " File not exist " + str(count))
o.close()
```
